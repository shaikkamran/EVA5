```python
import os
from glob import glob
import shutil
```

```python
x=glob('*')
```

```python

for i in x :
    
    if not os.path.isdir(i):
        continue
    
    li=glob(f"{i}/*")
    folder=f"{i}_renamed"
    if not os.path.exists(folder):
        os.mkdir (f"{folder}")
        
    for i,file in enumerate(li):

        dest_filename=f"{folder}/image_{str(i)}.jpg"
        shutil.copy(file,dest_filename)
```
